```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 22.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plot()
fid, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a level for the legend.
plt.plot(x_axis, y_axis, marker= '*', color='blue', linewidth=2, label='Boston')
# Create lavels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Set the y limit between 0 and 45. 
plt.ylim(0, 45)
# Create a title.
plt.title('PyBer Fare by Month')
# Add the legend.
plt.legend()
```

```python
# Create using the object orientated method.

ax = plt.axes()
ax.plot(x_axis, y_axis, color='green', linewidth=2, marker='D', label='Boston')
ax.legend()
ax.set_xlabel('Data')
ax.set_ylabel('Fare ($)')
ax.set_title('PyBer Fare by Month')
ax.set_ylim(0, 45)
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
fid, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c='skyblue', edgecolors='k', linewidths=2, alpha=.5, s=[i * 5 for i in y_axis], label='Boston')
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_xlim(0, 50)
ax.legend()
ax.invert_yaxis()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8, 8))

colors= ['slateblue', 'magenta', 'lightblue', 'green', 'yellowgreen', 'greenyellow', 'yellow', 'orange', 'gold', 'indianred', 'tomato', 'mistyrose']

plt.pie(y_axis,
 explode=explode_values,
  labels=x_axis,
   autopct='%.1f%%',
    colors=colors)

plt.show()

```

```python
fig, ax = plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode = (0,0,.5,0,0,0,.5,0,0,0,0,0)
ax.pie(y_axis,
 autopct='%.1f%%',
 labels=x_axis,
 colors = colors,
 explode=explode,
 shadow=True,
 startangle=270)
ax.invert_yaxis()
plt.show()
```
